```python
words=['apple', 'bat', 'bar', 'atom', 'book']
by_letter={}

for word in words:
    letter=word[0]
    by_letter.setdefault(letter, []).append(word)
```

```python
by_letter
```

```python
by_letter2={}

for word in words:
    letter=word[0]
    if letter not in by_letter2:
        by_letter2[letter]=[word]
    else:
        by_letter2[letter].append(word)
by_letter2        
```

```python
from collections import defaultdict
by_letter3={}
by_letter3=defaultdict(list)
for word in words:
    by_letter3[word[0]].append(word)
```

```python
by_letter3
```
